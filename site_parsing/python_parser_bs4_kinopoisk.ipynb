```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import lxml
# from unicodedata import normalize
```

## Парсинг

```python
URL = "https://www.kinopoisk.ru/lists/movies/top250/"

headers = {
    "Accept": "*/*",
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Mobile Safari/537.36"
}
req = requests.get(URL, headers=headers)
print(req.status_code)
```

```python
print(req.text)
```

```python
soup = bs(req.text, "lxml")
soup
```

```python
card = soup.find('a', class_='base-movie-main-info_link__YwtP1')
print(card)
```

```python
rate = soup.find('span', class_='styles_kinopoiskValue__9qXjg').text
rate
```

```python
films = soup.findAll('a', class_='base-movie-main-info_link__YwtP1')

data = []

for film in films:
    link = 'https://www.kinopoisk.ru' + film.get('href')
    local_name = film.find('div', class_='base-movie-main-info_mainInfo__ZL_u3').text
    try:
        original_name, year, length = film.find('span', class_='desktop-list-main-info_secondaryText__M_aus').text.replace(u'\xa0', u' ').split(', ')
    except:
        original_name = local_name
    data_group_parse = card.findAll('span', class_='desktop-list-main-info_truncatedText__IMQRP')
    bunch = data_group_parse[0].text.replace(u'\xa0\xa0', u' ').split(' ')
    country = bunch[0]
    film_type = bunch[2]
    director = bunch[4] + ' ' + bunch[5]
    cast = data_group_parse[1].text.replace('В ролях: ', '')
#     country_bunch = film.findAll('img', class_='countries_flag__5cI2E')
#     for c in country_bunch:
#         country = c.get('alt')
        
    
    data.append([link, local_name, original_name, year, length, film_type, director, cast, country])
#     data.append([country])
```
