```python
def extract_text_from_pdf(pdf_path: str, output_path: str) -> None:
    # Check if the PDF file exists
    if not Path(pdf_path).is_file():
        print(f"No such file: {pdf_path}")
        return

    # Open the PDF file in binary mode
    with open(pdf_path, "rb") as f:
        # Create a PdfFileReader object
        pdf_reader = PdfReader(f)

        # Get the number of pages in the PDF file
        num_pages = len(pdf_reader.pages)

        # Loop over the pages and extract the text
        for page_num in range(num_pages):
            # Get the page object
            page = pdf_reader.pages[page_num]

            # Extract the text from the page
            text = page.extract_text()

            # Write the text to the output file
            with open(output_path, "a", encoding="utf-8") as out_file:
                out_file.write(text)

    print(f"Text extracted from {pdf_path} and written to {output_path}")
```

```python
extract_text_from_pdf("train_dataset_Плантбокс/Атлас 2021.pdf",
                      "work_directory/Атлас 2021.txt")
extract_text_from_pdf("train_dataset_Плантбокс/Дикорастущие лекарственные растения СССР.pdf",
                      "work_directory/Дикорастущие лекарственные растения СССР.txt")
extract_text_from_pdf("train_dataset_Плантбокс/Лекарственные_и_эфирномасличные_растения_Маланкина_Е_,_Цицилин_А.pdf",
                      "work_directory/Лекарственные_и_эфирномасличные_растения_Маланкина_Е_,_Цицилин_А.txt")
```
