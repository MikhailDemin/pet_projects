```python
# Разделение по колонкам
def split_to_columns(text, column_list):
    dict_ = {}
    for i in range(len(column_list)-1):
#         dict_['Название'] = column_list[i]
        r = re.findall(rf'{(column_list[i])}(.*?){(column_list[i+1])}', text)
        if r:
            dict_[column_list[i]] = r
        else:
            print(column_list[i])
    last_pos = text.find(column_list[-1])
    dict_[column_list[-1]] = text[last_pos:]
    return dict_
```

```python
def make_table(plant_dict):
    keys = plant_dict.keys()
    plant_dict_all = []

    for i, key in enumerate(keys):
        abstracts = plant_dict[key]
        details = {}
        for j in range(len(abstracts)-1):
            row = ''.join(plant_data[key]).strip()

            start_pos = row.find(abstracts[j])
            end_pos = row.find(abstracts[j+1])
            details['Название.'] = key
            details[abstracts[j]] = row[start_pos+len(abstracts[j]):end_pos].strip()
            if j == (len(abstracts) - 1):
                start_pos = row.find(abstracts[j+1])
                details[abstracts[j+1]] = row[start_pos+len(abstracts[j+1]):].strip()

        plant_dict_all.append(details)
    return plant_dict_all
```

```python
abstracts = ['Описание.', 'Ареал.', 'Экология.', 'Ресурсы.', 'Химический состав.', 
             'Сырье.', 'Фармокологические свойства.', 'Применение в медицине.']
```

```python
plant_dict = {}
for i in plant_name:
    plant_dict[i] = abstracts
```

```python
plant_info_all = make_table(plant_dict)
```

```python
df = pd.DataFrame(plant_info_all)
```
