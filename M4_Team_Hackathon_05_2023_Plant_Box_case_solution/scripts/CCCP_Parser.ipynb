```python
import re

lines = []  # Empty list to store the lines

# Open the file in read mode using the 'with' statement
with open("Загрузки/Дикорастущие_лекарственные_растения_СССР.txt", "r") as file:
    lines = [line.strip() for line in file.readlines()]  # Read the lines and store them in the list

uppercase_russian_words = []
for line in lines:
    # Check if the word is uppercase and Russian
    if re.search(r'[А-ЯЁ]', line):
        uppercase_russian_words.append(line)  # Append the word to the list
uppercase_russian_words = [item for item in uppercase_russian_words if item[:3].isupper()]
uppercase_russian_words = [item for item in uppercase_russian_words if not item[:1].isdigit()]
first_two_words = [item.split()[:2] for item in uppercase_russian_words]
first_two_words2 = [i for i in first_two_words if re.search(r'[А-ЯЁ]', ' '.join(i))]
#print(uppercase_russian_words)  # Print the list of uppercase Russian words
```

```python
first_two_words2 = [i for i in first_two_words2 if i[0] not in ['В', '■', '♦', '|', '•', 'НО', 'I.', 'II.', 'IV.', 'Р', 'ЕРНЕ', 'КОЛЕНО', 'КОЛ']]
```

```python
a = []
for x in first_two_words2:
    a.append(' '.join(x))
```

```python
pattern = r'([а-яА-Я]+)\s*-\s*([а-яА-Я]+)'
# Define the regular expression pattern to match the text to be deleted
pattern2 = r'\(рис\.\s*\d+\)'

for i in range(len(a)-1):
    start_word = a[i]
    end_word = a[i+1]

    start_index = modified_text.find(start_word) + len(start_word)  # Find the position of the start word and add its length
    end_index = modified_text.find(end_word)  # Find the position of the end word

    extracted_text = modified_text[start_index:end_index].strip()  # Extract the text between the start and end words and remove any leading/trailing whitespace
    extracted_text = re.sub(pattern, r'\1\2', extracted_text)
    extracted_text = re.sub(pattern2, '', extracted_text)
    # print(start_word + ' ' + extracted_text)
    file_name = start_word + '.txt'  # Create the file name based on the start word

    with open(f'/home/mario/Загрузки/train_dataset_Плантбокс/test_PlanB/{file_name}', 'w') as output_file:  # Open the file for writing
        output_file.write(start_word + ' ' + extracted_text)  # Write the start word and extracted text to the file
```
